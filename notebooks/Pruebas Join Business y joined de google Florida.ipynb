```python
from geopy.distance import geodesic
import pandas as pd
from tqdm import tqdm

tqdm.pandas()


# Función para encontrar el negocio más cercano en df2 para cada entrada en df1
def find_closest_business(row, df2):
    min_distance = float('inf')
    closest_business = None
    for _, business in df2.iterrows():
        distance = geodesic((row['latitude'], row['longitude']), (business['latitude'], business['longitude'])).miles
        if distance < min_distance:
            min_distance = distance
            closest_business = business
    return closest_business['name'], min_distance

# Aplicar la función a cada fila en df1
df1[['closest_business', 'distance']] = df1.progress_apply(lambda row: find_closest_business(row, df2), axis=1, result_type='expand')

# Muestra el Dataframe actualizado
print(df1.head())
```

```python
from geopy.distance import geodesic
import math 
import numpy as np

def haversine_distance(lat1, lon1, lat2, lon2):
    # Convertir latitudes y longitudes a radianes (usando math.radians en vez de geopy.degrees)
    lat1_rad = math.radians(lat1)
    lon1_rad = math.radians(lon1)
    lat2_rad = math.radians(lat2)
    lon2_rad = math.radians(lon2)

    # Calcule el radio de la Tierra
    earth_radius = 6371  # km (or 3959 miles)

    # Aplicar la fórmula de Haversine
    delta_lat = lat2_rad - lat1_rad
    delta_lon = lon2_rad - lon1_rad
    a = np.sin(delta_lat / 2) ** 2 + np.cos(lat1_rad) * np.cos(lat2_rad) * np.sin(delta_lon / 2) ** 2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    # Calcular la distancia
    distance = earth_radius * c

    return distance
```

```python
# Calcular la latitud y longitud promedio
avg_lat = (df1['latitude'].mean() + df2['latitude'].mean()) / 2
avg_lon = (df1['longitude'].mean() + df2['longitude'].mean()) / 2

# Agregar columnas de distancia a los DataFrames
df1['distance_to_ref'] = df1.apply(lambda row: haversine_distance(row['latitude'], row['longitude'], avg_lat, avg_lon), axis=1)
df2['distance_to_ref'] = df2.apply(lambda row: haversine_distance(row['latitude'], row['longitude'], avg_lat, avg_lon), axis=1)
```

```python
# Unir los DataFrames basado en la distancia mínima
merged_df = df1.merge(df2, on='distance_to_ref', how='inner', sort=False)
```

```python
import geopandas as gpd
from shapely.geometry import Point
import pandas as pd

# Cargar los datasets
business_fl = pd.read_parquet('business_fl.parquet')
joined_restaurant_bars = pd.read_parquet('Joined_Restaurant_Bars_Florida_Google.parquet')

# Convertir dataframes a GeoDataFrames
business_fl_gdf = gpd.GeoDataFrame(business_fl, geometry=[Point(xy) for xy in zip(business_fl.longitude, business_fl.latitude)])
joined_restaurant_bars_gdf = gpd.GeoDataFrame(joined_restaurant_bars, geometry=[Point(xy) for xy in zip(joined_restaurant_bars.longitude, joined_restaurant_bars.latitude)])

# Establecer el sistema de referencia de coordenadas
business_fl_gdf.crs = 'EPSG:4326'
joined_restaurant_bars_gdf.crs = 'EPSG:4326'

# Unir espacialmente
result = gpd.sjoin(business_fl_gdf, joined_restaurant_bars_gdf, how='inner', op='intersects')

# Calcular la distancia entre los puntos
result['distance'] = result.apply(lambda row: row['geometry'].distance(Point(row['longitude_right'], row['latitude_right'])), axis=1)
filtered_result = result[result['distance'] <= 100]  # 1 km de distancia máxima

# Mostrar las primeras filas del dataset unido
print(filtered_result.head())
```

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors

# Cargar los datasets
business_fl = pd.read_parquet('business_fl.parquet')
joined_restaurant_bars = pd.read_parquet('Joined_Restaurant_Bars_Florida_Google.parquet')

# Extraer las coordenadas
coords_business = business_fl[['latitude', 'longitude']].values
coords_restaurants = joined_restaurant_bars[['latitude', 'longitude']].values

# Configurar el modelo de Nearest Neighbors
nbrs = NearestNeighbors(n_neighbors=1, algorithm='ball_tree').fit(coords_restaurants)
distances, indices = nbrs.kneighbors(coords_business)

# Agregar los índices y distancias al dataframe de negocios
business_fl['nearest_restaurant_index'] = indices.flatten()
business_fl['distance_to_nearest'] = distances.flatten()

# Hacer el merge basado en los índices de los vecinos más cercanos
merged_df5 = business_fl.merge(joined_restaurant_bars, left_on='nearest_restaurant_index', right_index=True, suffixes=('_business', '_restaurant'))

# Mostrar las primeras filas del dataframe unido
print(merged_df5.head())
print("Número de filas en merged_df:", merged_df5.shape[0])
```
