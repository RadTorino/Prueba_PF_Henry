```python
import pandas as pd
from textblob import TextBlob
from tqdm import tqdm

# Cargar el dataframe con la columna "estados"
df = pd.read_parquet('merged_starbucks_reviews_with_state.parquet')
```

```python


# Load the data
df = pd.read_parquet('merged_starbucks_reviews_with_state.parquet')

# Initialize tqdm for pandas
tqdm.pandas()

# Define a function to classify sentiment
def classify_sentiment(text):
    if pd.isna(text):
        return 1  # Treat NaN as Neutral
    analysis = TextBlob(str(text))
    if analysis.sentiment.polarity > 0:
        return 2  # Positive
    elif analysis.sentiment.polarity < 0:
        return 0  # Negative
    else:
        return 1  # Neutral

# Apply the function to the 'text' column
df['sentiment'] = df['text'].progress_apply(classify_sentiment)

# Display the head of the dataframe to show the new column
print(df.head())
```

```python
import pandas as pd
df_sentiment = pd.read_parquet('merged_starbucks_reviews_con_state_sentiment.parquet')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style('whitegrid')

# Count plot for sentiment
plt.figure(figsize=(10, 6))
sns.countplot(x='sentiment', data=df_sentiment, palette='viridis')
plt.title('Distribution of Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

Conclusión: Este gráfico confirma las observaciones del box plot, mostrando la densidad de las calificaciones para cada tipo de sentimiento. Las reseñas positivas (2) tienen una alta densidad de calificaciones altas, mientras que las reseñas negativas (0) tienen una alta densidad de calificaciones bajas. Las reseñas neutrales (1) muestran una distribución más dispersa de calificaciones.

```python
# Box plot for rating by sentiment
plt.figure(figsize=(10, 6))
sns.boxplot(x='sentiment', y='rating', data=df_sentiment, palette='viridis')
plt.title('Rating by Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Rating')
plt.show()
```

- Conclusión: La mayoría de las reseñas tienen un sentimiento positivo (2), seguido por las reseñas neutrales (1) y finalmente las reseñas negativas (0). Esto sugiere que, en general, los clientes tienen una percepción positiva de Starbucks.

```python
# Violin plot for rating by sentiment
plt.figure(figsize=(10, 6))
sns.violinplot(x='sentiment', y='rating', data=df_sentiment, palette='viridis')
plt.title('Rating Distribution by Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Rating')
plt.show()
```

Conclusión: Las calificaciones varían significativamente según el sentimiento. Las reseñas con sentimiento positivo (2) tienden a tener calificaciones más altas, mientras que las reseñas negativas (0) tienen calificaciones más bajas. Las reseñas neutrales (1) tienen una distribución de calificaciones más amplia, pero generalmente se encuentran en el rango medio.

```python
# Group by state and sentiment to get the count of each sentiment per state
state_sentiment_counts = df_sentiment.groupby(['state', 'sentiment']).size().unstack(fill_value=0)

# Normalize the counts to get the proportion of each sentiment per state
state_sentiment_proportions = state_sentiment_counts.div(state_sentiment_counts.sum(axis=1), axis=0)

# Plot the sentiment distribution by state
plt.figure(figsize=(15, 10))
state_sentiment_proportions.plot(kind='bar', stacked=True, colormap='viridis', figsize=(15, 10))
plt.title('Sentiment Distribution by State')
plt.xlabel('State')
plt.ylabel('Proportion')
plt.legend(title='Sentiment', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
