```python
import os
import pandas as pd

reviews = pd.read_parquet('wisconsin_reviews.parquet')
# Load the metadata file
metadata = pd.read_parquet('Metadata_Dunkin.parquet')
# Joinear los datasets en 'gmap_id'
merged_df_dunkin = pd.merge(metadata, reviews, on='gmap_id')

# mostrar el head of the dataframe joineado
merged_df_dunkin.head()
```

```python
number_of_rows = len(metadata)
print(number_of_rows)
```

```python
# Guardar el dataframe joineado en un nuevo archivo de Parquet
merged_df_dunkin.to_parquet('wisconsin_dunkin.parquet')
```
