```python
import pandas as pd
# Load the parquet file
df = pd.read_parquet('metadata-sitios.parquet')
# Display the head of the dataframe
df.head()
```

```python
# Filter the dataframe to include only rows where the 'name' column contains 'Dunkin'
df_dunkin = df[df['name'].str.contains('Dunkin', case=True, na=False)]
# Display the head of the filtered dataframe
df_dunkin.head()
```

```python
df_dunkin.to_parquet('Metadata_Dunkin.parquet')
```
