```python
import json

# Leer el archivo como texto sin formato para inspeccionar las primeras líneas
with open('1.json', 'r') as file:
    first_lines = [next(file) for _ in range(5)]

print("First lines of the file:\
", first_lines)

# Intentar cargar el archivo JSON como una lista de objetos JSON
with open('1.json', 'r') as file:
    data = []
    for line in file:
        try:
            data.append(json.loads(line))
        except json.JSONDecodeError as e:
            print("Error decoding JSON: ", e)
            break

# Convertir la lista de objetos JSON a un DataFrame
if data:
    df_1 = pd.DataFrame(data)
    print(df_1.head(30))
else:
    print("No data loaded.")
```

```python
import os

florida_reviews = pd.read_parquet('florida_reviews.parquet')

# Inicializar una lista vacía para almacenar datos de datos
dfs = []

# Cargar todos los archivos JSON y agregar a la lista
for i in range(1, 12):
    file_name = f'{i}.json'
    if os.path.exists(file_name):
        df = pd.read_json(file_name, lines=True)
        dfs.append(df)

# Concatene todos los dataframes de datos JSON
all_json_df = pd.concat(dfs, ignore_index=True)

# Filtrar el dataframe de datos JSON para el específico gmap_id
filtered_json_df = all_json_df[all_json_df['gmap_id'] == '0x88d8df729ca5eda7:0x8131504abdbfe7fe']

# joinear el parquet dataframe con el filtrado JSON dataframe en 'gmap_id'
merged_df = pd.merge(florida_reviews, filtered_json_df, on='gmap_id', how='inner')

# mostrar el head del join
merged_df.head(30)
```

```python
df1 = pd.read_parquet('1.parquet')
df2 = pd.read_parquet('2.parquet')
df3 = pd.read_parquet('3.parquet')
df4 = pd.read_parquet('4.parquet')
df5 = pd.read_parquet('5.parquet')
df6 = pd.read_parquet('6.parquet')
df7 = pd.read_parquet('7.parquet')
df8 = pd.read_parquet('8.parquet')
df9 = pd.read_parquet('9.parquet')
df10 = pd.read_parquet('10.parquet')
df11= pd.read_parquet('11.parquet')
df = pd.concat([df1, df2,df3,df4,df5,df6,df7,df8,df9,df10,df11], ignore_index=True)
# Manejar valores None en la columna 'category'
df['category'] = df['category'].apply(lambda x: x if x is not None else [])

# Filtrar el DataFrame para incluir solo filas donde la columna 'category' contiene 'Restaurant' o 'Bar'
filtered_df = df[df['category'].apply(lambda x: 'Restaurant' in x or 'Bar' in x)]

# Inspeccionar las primeras filas del DataFrame filtrado
filtered_df.head(10)
```

```python
# Guardar el DataFrame filtrado en un archivo parquet
filtered_df.to_parquet('Metdata_Restaurant_Bar.parquet')
print('El DataFrame filtrado se ha guardado como Metdata_Restaurant_Bar.parquet')
```
