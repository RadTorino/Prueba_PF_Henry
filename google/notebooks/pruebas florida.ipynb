```python
import pandas as pd
```

```python
# Cargar todos los datasets del 1 al 17
files = [str(i) + '.json' for i in range(1, 18)]
dfs = {}
for file in files:
    try:
        dfs[file] = pd.read_json(file, lines=True)
    except ValueError as e:
        print('Error al cargar el archivo', file, ':', e)

# Calcular la cantidad de información en cada dataset
info_counts = {file: dfs[file].count().sum() for file in dfs}

# Mostrar el dataset con más información
max_info_dataset = max(info_counts, key=info_counts.get)
print('El dataset con más información es:', max_info_dataset)
print('Cantidad de información:', info_counts[max_info_dataset])
```

```python
# Calcular la cantidad de valores nulos en cada dataset
null_counts = {file: dfs[file].isnull().sum().sum() for file in dfs}

# Mostrar el dataset con menos valores nulos
min_null_dataset = min(null_counts, key=null_counts.get)
print('El dataset con menos valores nulos es:', min_null_dataset)
print('Cantidad de valores nulos:', null_counts[min_null_dataset])
```

```python
import os

# lista de archivos json
json_files = ['1.json', '2.json', '3.json', '4.json', '5.json', '6.json', '7.json', '8.json', '9.json', '10.json', '11.json', '12.json', '13.json', '14.json', '15.json', '16.json', '17.json', '18.json', '19.json']

# leer y joinear todos los archivos json
all_dfs = []
for file in json_files:
    try:
        df = pd.read_json(file, lines=True) 
        all_dfs.append(df)
    except ValueError as e:
        print(f'Error reading {file}: {e}')

combined_df = pd.concat(all_dfs, ignore_index=True)
```

```python
# guardar a parquet
combined_df.to_parquet('florida_reviews.parquet')
```

```python
# Create a Boolean Series indicating rows containing 'Sweetgreen' (case-insensitive)
sweetgreen_matches = combined_df['name'].str.contains('Sweetgreen', case=False)

# Count the number of True values (occurrences)
sweetgreen_count = sweetgreen_matches.sum()

# Print the count
print(sweetgreen_count)
```
